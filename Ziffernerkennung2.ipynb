```python
def sigmoid(eingabe):
  return 1/(1+np.exp(-eingabe))
```

```python
def Abfrage(w2,b2,w3,b3,w4,b4, eingabe):
    z2=w2.dot(eingabe)+b2
    a2=sigmoid(z2)
    z3=w3.dot(a2)+b3
    a3=sigmoid(z3)
    z4=w4.dot(a3)+b4
    a4=sigmoid(z4)
    print(a4)
    return z2,a2,z3,a3,z4,a4
```

```python
def vorhersage_machen(a4):
    return np.argmax(a4,0)
```

```python
def make_predictions(X, W2, b2, W3, b3, w4, b4):
    _, _, _, _, _, A3 = Abfrage(W2, b2, W3, b3, w4, b4, X)
    predictions = vorhersage_machen(A3)
    return predictions
```

```python
def fileresize(filename):
  filepath1 = r'/content/drive/MyDrive/ABA Projekt Pickle/Ziffern/'
  ending=r'.jpg'
  file=filepath1+filename+ending
  test_image = cv2.imread(file, cv2.IMREAD_GRAYSCALE)
  plt.imshow(test_image, cmap='gray')
  img_resized = cv2.resize(test_image, (28,28), interpolation=cv2.INTER_LINEAR)
  img_resized = cv2.bitwise_not(img_resized)
  plt.imshow(img_resized, cmap='gray')
  return img_resized
```

```python
def Zifferneingabe(filename, W2, b2, W3, b3, W4, b4):
    bild=fileresize(filename)
    current_image = bild.reshape((784,1))/255
    prediction = make_predictions(current_image, W2, b2, W3, b3, w4, b4)
    current_image = current_image.reshape((28, 28)) * 255
    plt.gray()
    plt.imshow(current_image, interpolation='nearest')
    plt.show()
    print("Vorhersage der KI: ", prediction)
```
